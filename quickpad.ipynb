```python
import tkinter as tk
from tkinter import filedialog, Menu, messagebox, simpledialog

window = tk.Tk()
window.title("Untitled - Notepad")
window.geometry("500x500")

text = tk.Text(window, height=100, width=100)
text.pack()

def new_tab():
    text.delete(1.0, tk.END)  

def open_file():
    file = filedialog.askopenfilename()
    if file:  
        with open(file, 'r') as f:
            content = f.read()
            text.delete(1.0, tk.END) 
            text.insert(tk.END, content)  

def saveas_file():
    file = filedialog.asksaveasfilename(defaultextension=".txt",
                                          filetypes=[("Text files", "*.txt"),
                                                     ("All files", "*.*")])
    if file:
        with open(file, 'w') as f:
            f.write(text.get(1.0, tk.END))

def save_file():
    saveas_file()

def exit_app():
    if messagebox.askokcancel("Quit", "Do you really want to quit?"):
        window.quit()

def cut_text():
    text.event_generate("<<Cut>>")

def copy_text():
    text.event_generate("<<Copy>>")

def paste_text():
    text.event_generate("<<Paste>>")

def delete_text():
    try:
        text.delete("sel.first", "sel.last")
    except tk.TclError:
        messagebox.showinfo("Info", "No text selected to delete.")

def replace_text():
    find_string = simpledialog.askstring("Find", "Enter the text to find:")
    if find_string is None:
        return  

    replace_string = simpledialog.askstring("Replace", "Enter the replacement text:")
    if replace_string is None:
        return  

    content = text.get(1.0, tk.END)
    new_content = content.replace(find_string, replace_string)
    text.delete(1.0, tk.END)
    text.insert(tk.END, new_content)

menu = Menu(window)
file = Menu(menu, tearoff=0)
file.add_command(label="New tab", command=new_tab)  
file.add_command(label="Open", command=open_file)  
file.add_command(label="Save", command=save_file)
file.add_command(label="Save as", command=saveas_file)
file.add_separator()
file.add_command(label="Exit", command=exit_app)

edit = Menu(menu, tearoff=0)
edit.add_command(label="Cut", command=cut_text)
edit.add_command(label="Copy", command=copy_text)
edit.add_command(label="Paste", command=paste_text)
edit.add_command(label="Delete", command=delete_text)
edit.add_separator()
edit.add_command(label="Replace", command=replace_text)

view = Menu(menu, tearoff=0)
view.add_command(label="Zoom in")
view.add_command(label="Zoom out")

menu.add_cascade(label="File", menu=file)
menu.add_cascade(label="Edit", menu=edit)
menu.add_cascade(label="View", menu=view)

window.config(menu=menu)

window.mainloop()
```
